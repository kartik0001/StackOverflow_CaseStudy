## Stack Overflow 2018 Developer Survey

### Analysis of Survey of StackOverflow in all over the world on the basis of Countries and Developer Type

```python
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('survey_results_public.csv')
```

```python
df1 = df.Country.value_counts().nlargest(20)
df2 = pd.DataFrame({'country':df1.index, 'count':df1.values})
f, ax = plt.subplots(figsize=(20, 10))
sns.barplot(y='country',x='count', data=df2).set_title('Top 20 countries who participated')
```

#### The above graph represents Top 20 countries who participated in StackOverflow on daily basis, from above graph we concluded that more than 20,000 people of US contributed to Opensource and around 14,000 people from INDIA contributed daily on Opensource.

```python
df3 = df[['Respondent','Country','Gender']]
df4 = df3.loc[df['Gender']=='Male']
df5 = df4.Country.value_counts().nlargest(20)
df6 = pd.DataFrame({'Country':df5.index, 'Count':df5.values})
df6.head()
f, ax = plt.subplots(figsize=(20, 10))
sns.barplot(y='Country',x='Count', data=df6).set_title('Top 20 countries with Male respondents')
```

#### The above graph represents Top 20 countries with Male who has responded to the queries that has been put up in StackOverflow on daily basis.Approx.. 14,000 male responded from US and approx..6,500 male responded from INDIA.  

```python
df3 = df[['Respondent','Country','Gender']]
df4 = df3.loc[df['Gender']=='Female']
df5 = df4.Country.value_counts().nlargest(20)
df6 = pd.DataFrame({'Country':df5.index, 'Count':df5.values})
df6.head()
f, ax = plt.subplots(figsize=(20, 10))
sns.barplot(y='Country',x='Count', data=df6).set_title('Top 20 countries with Female respondents')
```

#### The above graph represents Top 20 countries with Female who has responded to the queries that has been put up in StackOverflow on daily basis.Approx.. 13,00 female responded from US and approx..450 female responded from INDIA.  

```python
df1 = df['DevType'].str.split(';',expand=True)
df2=df1.fillna('')
df3 = df2.stack().value_counts().nlargest(20)
df4 = pd.DataFrame({'DevType':df3.index, 'Count':df3.values})
df5 = df4.iloc[1:]
f, ax = plt.subplots(figsize=(20, 10))
sns.barplot(y='DevType',x='Count', data=df5).set_title('Top 20 Developer Types')
```

#### While reading the above graph we come up to a solution that there are near about 58% profiles are Back-end Developer ,48% profiles are Full Stack Developer , 38% profiles are Front-End Developer , 21% profiles are Mobile Developer and for further info you can read the above graph.
